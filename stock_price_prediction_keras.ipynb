```python
import tensorflow as tf
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
```

```python
dataset = np.loadtxt('datasets/normalized_apple_prices.csv')
dataset = dataset[::-1]
```

```python
plt.plot(dataset)
plt.show()
```

```python
# Tailor the input

def transform_series(series, window_size):
    x = list()
    y = list()
    
    l = len(series)
    i = 0
    
    while (i+window_size) < l:
        x.append(series[i:i+window_size])
        y.append(series[i+window_size])
        i += 1
    
    return x, y
```

```python
window_size = 5
x, y = transform_series(dataset, window_size)
```

```python
# Split the dataset

test_train_split = int(np.ceil(len(y)*2/float(3)))

x_train = np.array(x[:test_train_split])
y_train = np.array(y[:test_train_split])

x_test = np.array(x[test_train_split:])
y_test = np.array(y[test_train_split:])

```

```python
x_train = x_train.reshape((x_train.shape[0], window_size, 1))
x_test = x_test.reshape((x_test.shape[0], window_size, 1))
```

```python
# The model

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.optimizers import RMSprop

model = Sequential()
model.add(LSTM(5, input_shape=(window_size, 1)))
model.add(Dense(1))

opt = RMSprop()
model.compile(loss='mean_squared_error', optimizer=opt)
```

```python
model.fit(x_train, y_train, epochs=1000, batch_size=50)
```

```python
train_predict = model.predict(x_train)
test_predict = model.predict(x_test)
```

```python
model.evaluate(x_train, y_train, verbose=0)
```

```python
model.evaluate(x_test, y_test, verbose=0)
```

```python
fig = plt.figure(figsize=(15, 10))
plt.plot(dataset, c='k')

split_pt = test_train_split + window_size
plt.plot(np.arange(window_size, split_pt, 1), train_predict, c='b')
plt.plot(np.arange(split_pt, split_pt + len(test_predict), 1), test_predict, c='r')

plt.xlabel('day')
plt.ylabel('Normalized price of Apple stock')
plt.legend(['original', 'training fit', 'test fit'])
plt.show()
```
